ジグザグGNRのバンド計算

```python
#GNRのバンド計算
import numpy as np
import matplotlib.pyplot as plt

N=50 # number of sites
t=1 # hopping parameter

def cal_band_zigzag(k):
    H=np.zeros((2*N,2*N),dtype=complex)
    for i in range(N):
        idx=2*i
        idx_plus=2*(i+1)
        H[idx+1,idx]=-t*(np.exp(-1j*k)+1)
        if idx_plus<2*N:
            H[idx_plus,idx+1]=-t
    H+=np.conjugate(H).T

        
    eig=np.linalg.eigvalsh(H)
    return eig

k_list=np.linspace(0,2*np.pi,1000)
eig=np.array([cal_band_zigzag(k) for k in k_list])
new_k_list=np.linspace(0,2,1000)

plt.plot(new_k_list,eig,color="black",linewidth=0.5)
plt.ylabel("E/t")
plt.xlabel("$k_ya/\\pi$")
plt.title("band structure of zigzag GNR (N={})".format(N))
plt.show()   
```

Nを変化させながらアニメーションを表示

```python
#GNRのバンド計算
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

t=1 # hopping parameter

fig, ax = plt.subplots()
ims=[]
N_list=np.arange(1,51)
def cal_band_zigzag(k,N):
    H=np.zeros((2*N,2*N),dtype=complex)
    for i in range(N):
        idx=2*i
        idx_plus=2*(i+1)
        H[idx+1,idx]=-t*(np.exp(-1j*k)+1)
        if idx_plus<2*N:
            H[idx_plus,idx+1]=-t
    H+=np.conjugate(H).T       
    eig=np.linalg.eigvalsh(H)
    return eig

k_list=np.linspace(0,2*np.pi,1000)
for j in N_list:
    eig=np.array([cal_band_zigzag(k,j) for k in k_list])
    new_k_list=np.linspace(0,2,1000)
    ax.set_ylabel("E/t")
    ax.set_xlabel("$k_ya/\\pi$")
    # ax.set_title("band structure of zigzag GNR (N={})".format(j))
    title = ax.text(0.5, 1.01, 'band structure of zigzag GNR (N={})'.format(j),ha='center', va='bottom',transform=ax.transAxes, fontsize='large')
    im=ax.plot(new_k_list,eig,color="black",linewidth=0.5)
    ims.append(im + [title])
ani = animation.ArtistAnimation(fig, ims,interval=400)
ani.save('zigzag_GNR.gif', writer='imagemagick')
```

アームチェアGNRのバンド構造

```python
#GNRのバンド計算
import numpy as np
import matplotlib.pyplot as plt

N=5 # number of sites
t=1 # hopping parameter

def cal_band_armchair(k):
    H=np.zeros((2*N,2*N),dtype=complex)
    for i in range(N):
        idx=2*i
        idx_plus=2*(i+1)
        if i%2==0:
            H[idx+1,idx]=-t*(np.exp(1j*k))
        if i%2==1:
            H[idx+1,idx]=-t
        if idx_plus<2*N:
            H[idx_plus,idx+1]=-t
            H[idx_plus+1,idx]=-t
    H+=np.conjugate(H).T    
    eig=np.linalg.eigvalsh(H)
    return eig

k_list=np.linspace(-np.pi,np.pi,1000)
eig=np.array([cal_band_armchair(k) for k in k_list])

plt.plot(k_list,eig,color="black",linewidth=0.5)
plt.ylabel("E/t")
plt.xlabel("$k_ya$")
plt.title("band structure of armchair GNR (N={})".format(N))
plt.show()   
```

```python
#GNRのバンド計算
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

t=1 # hopping parameter

fig, ax = plt.subplots()
ims=[]
N_list=np.arange(1,51)
def cal_band_armchair(k,N):
    H=np.zeros((2*N,2*N),dtype=complex)
    for i in range(N):
        idx=2*i
        idx_plus=2*(i+1)
        if i%2==0:
            H[idx+1,idx]=-t*(np.exp(1j*k))
        if i%2==1:
            H[idx+1,idx]=-t
        if idx_plus<2*N:
            H[idx_plus,idx+1]=-t
            H[idx_plus+1,idx]=-t
    H+=np.conjugate(H).T    
    eig=np.linalg.eigvalsh(H)
    return eig

k_list=np.linspace(-np.pi,np.pi,1000)
for j in N_list:
    eig=np.array([cal_band_armchair(k,j) for k in k_list])
    new_k_list=np.linspace(-1,1,1000)
    ax.set_ylabel("E/t")
    ax.set_xlabel("$k_ya/\\pi$")
    # ax.set_title(f"Band structure of armchair GNR (N={j})")
    title = ax.text(0.5, 1.01, 'band structure of armchair GNR (N={:.2f})'.format(int(j)),ha='center', va='bottom',
                    transform=ax.transAxes, fontsize='large')
    im=ax.plot(new_k_list,eig,color="black",linewidth=0.5)
    ims.append(im + [title])
ani = animation.ArtistAnimation(fig, ims,interval=200)
ani.save('armchair_GNR.gif', writer='imagemagick')
```
